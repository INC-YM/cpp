2. [0,1]之间有多少小数。无穷个

但是计算机的bit是有限的，所以为了尽可能多地表示小数，计算会用一个中间数表示其相邻的几个小数

3. 浮点数会带来微小的误差，我们要做的就是管理误差

4. 浮点数的格式

5. double
能够有64位bits表示

6. long double <br>

7. 浮点数的优势

因为有指数，所以浮点数能表示的范围比整数大很多<br>
运算速度比整数慢<br>
