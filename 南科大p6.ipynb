1. 变量必须初始化

2. 输出结果与实际计算不符合
因为结果的二进制超过了32位，而int最高只能表示到32位，且最高位要用来表示符号

3. signed and unsigned
有符号的整数，最高位用来表示±,-2^31~(2^31-1),如果二进制是32位
无符号则全部可以用来表示数字0~(2^32-1)，如果二进制是32位


4. long int 获取更长的整数
5. short int 获取更短的整数
6. long long int 更更长
但C C++ 标准没有固定上面的位数
**然后C++新增一个特性，规定long long int 为64位**

**7. 操作符sizeof**
注意，它不是函数
sizeof(int) 是返回该int占的字节
